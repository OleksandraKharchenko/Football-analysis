```python
import requests
import json
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/statsbomb/open-data/master/data/competitions.json"

response = requests.get(url)

if response.status_code == 200:
    json_data = response.json()
    df = pd.DataFrame(json_data)
    # Process the DataFrame here
    df.head()
else:
    print("Error:", response.status_code)
```

```python
# Printing all matches
with open('open-data-master/open-data-master/data/matches/43/106.json') as f:
    data = json.load(f)
    for match in data:
        print('ID:', match['match_id'], match['home_team']['home_team_name'], match['home_score'], '-', match['away_score'], match['away_team']['away_team_name'], ", date:", match['match_date'], "time:" , match['kick_off'])
```

```python
with open('open-data-master/open-data-master/data/matches/43/106.json') as f:
    data = json.load(f)
    
    matches = []
    for match in data:
        match_info = {
            'ID': match['match_id'],
            'Home Team': match['home_team']['home_team_name'],
            'Home Score': match['home_score'],
            'Away Score': match['away_score'],
            'Away Team': match['away_team']['away_team_name'],
            'Date': match['match_date'],
            'Time': match['kick_off']
        }
        matches.append(match_info)
    
    df = pd.DataFrame(matches)
    df['Datetime'] = pd.to_datetime(df['Date'] + ' ' + df['Time'])
    df = df.sort_values('Datetime')
    df.head()
```
